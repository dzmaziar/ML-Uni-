```python
data['status'] = data.status.map({'acquired':1, 'closed':0})
data
```

```python
data.corr()
```

```python

data=data.drop(data[data.age_first_funding_year<0].index)
data=data.drop(data[data.age_last_funding_year<0].index)
data=data.drop(data[data.age_first_milestone_year<0].index)
data=data.drop(data[data.age_last_milestone_year<0].index)
```

```python
data.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
features = ['age_first_funding_year', 'age_last_funding_year','age_first_milestone_year', 
               'age_last_milestone_year', 'relationships','funding_rounds', 'funding_total_usd',
               'milestones','avg_participants']
plt.figure(figsize=(15, 7))
for i in range(0, len(features)):
    plt.subplot(1, 11, i+1)
    sns.boxplot(y=data[features[i]],color='green',orient='v')
    plt.tight_layout()
```

```python
import numpy as np
from sklearn.preprocessing import MinMaxScaler, StandardScaler
buffs= ['funding_total_usd','relationships','avg_participants','age_first_funding_year']
for var in buffs :
    data['norm_'+var] =np.log1p(data[var])

buffs_2=['norm_funding_total_usd','norm_age_first_funding_year', 'norm_relationships','norm_avg_participants']

for var in buffs_2 :
    data[var] = MinMaxScaler().fit_transform(data[var].values.reshape(len(data), 1))
```

```python
plt.figure(figsize=(15, 7))
for i in range(0, len(buffs_2)):
    plt.subplot(1, 12, i+1)
    sns.boxplot(y=data[buffs_2[i]],color='green',orient='v')
    plt.tight_layout()
```

```python
for_delete = ['state_code','id','Unnamed: 6','category_code','object_id','zip_code','founded_at','closed_at','first_funding_at','last_funding_at','city','name','Unnamed: 0','latitude','longitude']

data=data.drop(for_delete,axis=1)

data.head()
```

```python
data=data.drop(['labels'],axis=1)
data.head()
```

```python
data=data.drop(['state_code.1'],axis=1)
data.head()
```

```python
data.info()
```

```python
X = data.drop(['status'],axis=1)

Y = data['status']

```

```python
Y
```

```python
X['age_first_milestone_year']=data['age_first_milestone_year'].fillna(0)
X['age_last_milestone_year']=data['age_last_milestone_year'].fillna(0)
```

```python
from sklearn.model_selection import train_test_split 
from imblearn import under_sampling, over_sampling
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
from sklearn.metrics import roc_auc_score

X_train, X_test,y_train,y_test = train_test_split(X,
                                                Y,
                                                test_size = 0.2,
                                              random_state = 42)

model = GaussianNB()

model.fit(X_train,y_train)
answer=model.predict(X_test)


print('\naccuracy')
print(accuracy_score(y_test, answer))

roc_auc_score(y_test,answer)
print('ROC Score:',roc_auc_score(y_test, answer))

print("Train Accuracy:",model.score(X_train, y_train))
print("Test Accuracy:",model.score(X_test, y_test))

```

```python
from sklearn.ensemble import RandomForestClassifier
model2=RandomForestClassifier()
model2.fit(X_train,y_train)
answer2=model2.predict(X_test)

print('\naccuracy')
print(accuracy_score(y_test, answer2))

roc_auc_score(y_test,answer2)
print('ROC Score:',roc_auc_score(y_test, answer2))

print("Train Accuracy:",model2.score(X_train, y_train))
print("Test Accuracy:",model2.score(X_test, y_test))
```

```python
from sklearn.svm import SVC
model3 = SVC()

model3.fit(X_train,y_train)


answer3 = model3.predict(X_test)

print('\naccuracy')
print(accuracy_score(y_test, answer3))

roc_auc_score(y_test,answer3)
print('ROC Score:',roc_auc_score(y_test, answer3))

print("Train Accuracy:",model3.score(X_train, y_train))
print("Test Accuracy:",model3.score(X_test, y_test))
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, answer)
plt.rcParams['figure.figsize'] = (3, 3)
sns.heatmap(cm, annot = True, cmap = 'YlGnBu', fmt = '.8g')
plt.show()

cm2 = confusion_matrix(y_test, answer2)
plt.rcParams['figure.figsize'] = (3, 3)
sns.heatmap(cm2, annot = True, cmap = 'YlGnBu', fmt = '.8g')
plt.show()

cm3 = confusion_matrix(y_test, answer3)
plt.rcParams['figure.figsize'] = (3, 3)
sns.heatmap(cm3, annot = True, cmap = 'YlGnBu', fmt = '.8g')
plt.show()
```

```python
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score

print('precision - сколько классификатор назвал положительными, recall - сколько из них реально положительные')
scores = {
                  'SVC':  { 
                             'precision_score': precision_score(y_test, answer3),
                             'recall_score': recall_score(y_test, answer3)
                         },
                 'Naive Bayes':  { 
                             'precision_score': precision_score(y_test, answer),
                             'recall_score': recall_score(y_test, answer)
                         },
                 'Random Forest':  { 
                             'precision_score': precision_score(y_test, answer2),
                            'recall_score': recall_score(y_test, answer2)
                         }
            }
  

scores = pd.DataFrame(scores)
scores.plot(kind="barh",figsize=(12, 12)).legend(loc='upper center', ncol=3, title="Результаты")
```
